```python
# Packages we need

import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
v_totales = pd.read_csv('Data/ventas_totales.csv', encoding='utf-8', sep=';', thousands='.')
v_totales.head(7)
```

```python
v_totales.set_index(['año', 'sector', 'size'], inplace=True)
v_totales = v_totales.stack()
v_totales = v_totales.reset_index()
v_totales.head(10)

# Reformar el marco de datos: https://pandas.pydata.org/pandas-docs/stable/user_guide/reshaping.html
```

```python
v_totales = v_totales.rename(columns={'size': 'Tamaño de la empresa', 'level_3': 'Cod_cant', 0: 'Ventas totales'}, inplace = False)
v_totales.head(5)
```

```python
v_totales = v_totales.groupby(by=['año', 'Cod_cant', 'sector']).sum().reset_index()
```

```python
v_totales['sector'].values
```

```python
v_totales = v_totales[v_totales['sector'] == 'Servicios']
```

```python
# Load maps

mapa = gpd.read_file('mapa_ecuador/', layer='ecu_admbnda_adm2_inec_20190724')
mapa.head(6)
```

```python
# En este caso, es conveniente unir las bases por códigos de cantón para no perder información. 
# Hay que hacer un corte del strung para eso.

mapa['Cod_cant'] = mapa['ADM2_PCODE'].str.slice(start=2) # Los códigos de cantón están contenidos en ADM2_ES
```

```python
v_totales = mapa.merge(v_totales, on=['Cod_cant'])
```

```python
v_totales.columns
```

```python
v_totales.head(10)
```

# Dibujar mapas

```python
import matplotlib.pyplot as plt
```

```python
v_totales['% del total ventas servicios'] = v_totales['Ventas totales'].apply(lambda row: 100*row/v_totales['Ventas totales'].sum())
```

```python
v_totales_wg = v_totales[v_totales['ADM1_ES'] != 'Galapagos']
v_totales_wg = v_totales_wg[v_totales_wg['año'] != '2020']
```

```python
v_totales_wg.columns
```

```python
v_totales_wg = v_totales_wg[['Shape_Leng', 'Shape_Area', 'geometry', 'Ventas totales', '% del total ventas servicios']]
```

```python
# Setting , like decimal separator

pd.options.display.float_format = '{:,.2f}'.format
```

```python
# Crear ejes con contenedor y figura personalizando tamañp

fig, ax = plt.subplots(figsize=(10, 10))
 
# Mostrar el mapa finalizado

v_totales_wg.plot(column='Ventas totales',
        ax=ax, zorder=2,
       )

# Remover cuadros

ax.axis('off')


# TITLES

ax.set_title('Participación de los servicios en la economía, 2020', 
             fontdict={'fontsize': '25', 'fontweight' : '3'})

# Create an annotation for the data source

ax.annotate('Fuente: BCE y OCHA\nAutor: Michael Peñafiel',
            xy=(0.1, .08), 
            xycoords='figure fraction', 
            horizontalalignment='left', 
            verticalalignment='top', 
            fontsize=15, 
            color='#555555'
           )

#Definition of the graphhical bar

bar = plt.cm.ScalarMappable(norm=plt.Normalize(vmin = v_totales_wg['Ventas totales'].min(), 
                                               vmax=v_totales_wg['Ventas totales'].max()
                                              )
                           )

#It is declared a list where the values of the bar will be saved

bar._A = []

#Location and size of the bar

cax = plt.axes([0.85, 0.15, 0.03, 0.7]) #[xcoord, ycoord, width, long]

# Plot of the graphical bar with the upper variables (bar and cax)

cbar = fig.colorbar(bar, cax=cax)

#Bar title

cbar.set_label('Ventas totales (miles de dólares)', fontsize=15)

# Guardar imagen 

plt.savefig('Graphs/ventas.png', dpi=300)
```
