```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import os
import scipy as sp
```

```python
# Identificación de los archivos/file identification

os.listdir('Edu1/')
```

```python
# Cargar datos/Load data
# low_memory es un parámetro para indicar que en las columnas del marco de datos/data frame (df) hay varios tipos de valores (floats, ints, strings and other)

df = pd.read_csv('Edu1/SBAC19_micro_514852_20200325_CSV.zip', sep=';', low_memory=False, decimal=",")
```

```python
# Mostrar datos/Show data

df.head(10)
```

```python
# Estudiar la variables

for i in df.columns:
    print(i)
```

## Muestro

```python
"""Los datos utilizados expresan una heterogenidad interna en gsubgrupos a tratar, por lo que se ha usado conglomerados."""

# Para recuperar una muestra hay que vincular algunas variables y luego seleccionar aleatoriamente observaciones
# Also we need select scolar population
df[df['poblacion'] == 1]
df = df.groupby(['id_dist', 'id_circ', 'id_prov', 'id_cant', 'id_parr', 'financiamiento', 'tp_area'])
df_r = df.sample(frac=0.30, replace=True, random_state=1)
df_r.head(7)
```

```python
# Existe un problema más: la existencia de datos perdidos en la base de datos en las variables que queremos tratar
# 999999 codifica los valores perdidos
df_r = df_r[df_r['isec'] != 999999]
df_r = df_r[df_r['inev'] != 999999]
df_r = df_r.dropna(subset=['isec', 'inev']) # Finalmente quitamos los nans
df_r.head(10)
```

### Grafiquemos una correlación útil

```python
fig, ax = plt.subplots(figsize=(6, 6)) # Para crear un fondo y los ejes respectivamente
df_r['isec']
colors = {'1':'red', '2':'green', '3':'blue', '4':'yellow', '5':'pink'}
ax.scatter(np.log2(df_r['isec'].abs()), np.log2(df_r['inev']), c=df_r['etnibbe'].astype('string').map(colors))

plt.show()

# Reference: https://kanoki.org/2020/08/30/matplotlib-scatter-plot-color-by-category-in-python/
```

```python
grid = sns.FacetGrid(df_r, col="etnibbe", hue="quintil", col_wrap=5)
grid.map(sns.regplot, "isec", "inev")

grid.add_legend()

plt.show()

# Source: https://www.geeksforgeeks.org/python-seaborn-facetgrid-method/
```

```python
grid = sns.FacetGrid(df_r, col='quintil', hue='etnibbe', col_wrap=5)
grid.map(sns.regplot, "isec", "inev")

grid.add_legend()

plt.show()
```

```python
grid = sns.FacetGrid(df_r, col='etnibbe', row='financiamiento')
grid.map(sns.regplot, "isec", "inev", scatter_kws={'alpha': 0.4}, line_kws={'color': 'black'})

def annotate(data, **kws):
    r, p = sp.stats.pearsonr(df_r['isec'], df_r['inev'])
    ax = plt.gca()
    ax.text(.05, .8, 'r={:.2f}, p={:.2g}'.format(r, p),
            transform=ax.transAxes)

grid.map_dataframe(annotate)
grid.add_legend()

plt.show()

# Source: https://www.geeksforgeeks.org/python-seaborn-facetgrid-method/
# Data of r: https://stackoverflow.com/questions/25579227/seaborn-implot-with-equation-and-r2-text
# More
```
