```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
# Datas

mapa = gpd.read_file('Data\cantones-ecuador.geojson')
df = pd.read_csv('Data\data.csv', sep=';', low_memory=False)
```

```python
# La base del INEC no tiene los codes likes strings, por desgracia.

mapa['DPA_CANTON'] = mapa['DPA_CANTON'].astype(int)
mapa['DPA_PROVIN'] = mapa['DPA_PROVIN'].astype(int)
mapa.head(100)
```

```python
df_merged = pd.merge(mapa, df, right_on=['id_prov', 'id_cant'], left_on=['DPA_PROVIN', 'DPA_CANTON'], how='left')
df_merged.shape
```

```python
df_merged = df_merged[['geometry', 'etnibbe']]
df_merged =  df_merged.dropna()
df_merged['etnibbe'] = df_merged['etnibbe'].astype(int)
df_merged.head(10)
```

```python
type(df_merged)
```

```python
fig, ax = plt.subplots(1, figsize=(15, 15))
df_merged['geometry'].plot(ax=ax)
```
