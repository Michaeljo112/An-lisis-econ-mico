# Share of servicies in Latinamerica & Caribbean

```python
# Read packages

import pandas as pd
import numpy as np
```

```python
# Load datasets and dictionaries

df = pd.read_table('Data/dc3cefcf-3b8b-46ba-b024-a8398c49bf2e_Data.txt', decimal='.', index_col=False)
df.head(10)
```

```python
meta = pd.read_csv('Data/Metadata_Country_API_NV.SRV.TOTL.ZS_DS2_en_csv_v2_3359028.csv', sep=',')
meta.head(10)
```

```python
# To asses columns names

df.columns
```

```python
# Select interets' years

df = df[['Country Name', 'Country Code', '2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]']]
```

```python
# Merge datas (regions with counstries)

df = df.merge(meta, on=['Country Code'])
```

```python
# Select only Latin America & Caribbean

df_la = df[df['Region'] == 'Latin America & Caribbean']
df_la.sort_values(['Country Code'], ascending=True)
```

```python
# Due to exist a problem with no data codifing like '..', we need transform data

df_la = df_la.replace(to_replace="..", value='nan')
```

```python
df_la
```

```python
# Transfor columns in floats columns

L = ['2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]']
for year in L:
    df_la[year] = df_la[year].astype(float) 
```

```python
# I prefer use data of 2019

df_la = df_la.dropna(subset=['2020 [YR2020]'])
df_la = df_la.sort_values(['Country Code'], ascending=True)
```

# Plots of data

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams["figure.autolayout"] = True

# Create figures

fig, ax = plt.subplots(figsize=(10, 8))
plot = plt.bar(x='Country Code', height='2020 [YR2020]', data=df_la.sort_values(['2020 [YR2020]'], ascending=False)) # rwidth aumenta el tamaño del espacio entre las barras

# Add the data value on head of the bar

for value in plot:
    height = value.get_height()
    plt.text(value.get_x() + value.get_width()/2.,
             1.002*height, round(value.get_height(), 1), ha='center', va='bottom',
             fontsize=10
            )

plot[22].set_color('red')

# Ajustar estilo

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')
ax.tick_params(bottom=False, left=False)
ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)

# Add titles

plt.title("Latin America & Caribbean. 2020 services' share", fontsize=20)
plt.annotate('Source: World Bank\nAutor: Michael Peñafiel',
            xy=(0.1, .06), 
            xycoords='figure fraction', 
            horizontalalignment='left', 
            verticalalignment='top', 
            fontsize=15, 
            color='#555555')
plt.xticks(horizontalalignment="center", fontsize=13, rotation=90)
plt.yticks(fontsize=16)
plt.ylabel("% GDP", fontsize=16)
plt.savefig('Graphs/1.png', dpi=300)

# For more detail read https://stackoverflow.com/questions/18973404/setting-different-bar-color-in-matplotlib-python
```
