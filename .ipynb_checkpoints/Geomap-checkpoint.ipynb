```python
import pandas as pd 
import geopandas as gpd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
from unicodedata import normalize # Nos ayuda a recodificar valores unicode, utf-8 a latin
```

```python
# Leemos los datos por límites administrativos a nivel cantonal

mapa = gpd.read_file('mapa', layer='ecu_admbnda_adm3_inec_20190724')
```

```python
# Identificamos la dimensión de la base de datos

mapa.shape
```

```python
# Revisamos los nombres de las variables

mapa.columns
```

```python
mapa.head(100)
```

```python
# Debido a que los nombres de las provincias están escritos de manera diferentes en las dos bases de datos a tratar, 
# por ejemplo un Cantón con una ñ o una palabra con tilde en una de las dos bases está escrito sin ñ o tilde. 
# Para no perder información, vamos a codificar las palabras en latin, y a convertirlas en mayúsculas y ganar coincidencias.
# Vamos a empezar creando una lista que pueda reemplazar a las columnas originales con tildes y caracteres que quitar 
```

```python
# COdificar/quitar tildes

L = ['ADM1_ES', 'ADM2_ES', 'ADM3_ES']
for i in L:
    R = []
    for canton in mapa[i].str.upper():
        c = normalize('NFKD', canton).encode('latin', 'ignore').decode('latin')
        R.append(c)
    mapa[i] = R
```

```python
# Verificamos el reemplazo

tup = []
for ref in L:
    tup.append(len(mapa[ref]))
tup

# El número de filas de mapa debería repetirse 3 veces
```
