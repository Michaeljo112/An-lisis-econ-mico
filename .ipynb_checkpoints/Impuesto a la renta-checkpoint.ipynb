```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
from unicodedata import normalize # Nos ayuda a recodificar valores unicode, utf-8 a latin
```

```python
# Setting , like decimal separator

pd.options.display.float_format = '{:,.2f}'.format
```

```python
df = pd.read_excel('Data\sa.xlsx', sheet_name=1)
mapa = gpd.read_file('mapa', layer='ecu_admbnda_adm3_inec_20190724')
```

```python
mapa.head()
```

```python
mapa['ADM1_ES'] = mapa['ADM1_ES'].str.upper()
mapa['ADM2_ES'] = mapa['ADM2_ES'].str.upper()
```

```python
df_merged = mapa.merge(df, right_on=['Provincia', 'Cantón'], left_on=['ADM1_ES', 'ADM2_ES'], how='left')
```

```python
df_merged = df_merged[df_merged['Provincia'] != 'GALAPAGOS'] 
df_merged.head()
```

```python
fig, ax = plt.subplots(1, figsize=(10, 10))

# Control del título y los ejes

ax.set_title('Recaudación del impuesto a la renta', 
             pad=60, 
             fontdict={'fontsize':20, 'color': 'black'})

# Obetener mapa

df_merged.plot(column='recaudación_2020', cmap='plasma', ax=ax,
         zorder=2, legend=True,
          legend_kwds={'label': "Recaudación del impuesto a la renta",
                        'orientation': "horizontal"})
ax.axis('off')

plt.savefig('red.png', dpi=500)
```
