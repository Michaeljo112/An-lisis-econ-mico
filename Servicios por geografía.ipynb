```python
# Packages we need

import pandas as pd
import numpy as np
import geopandas as gpd
```

# Participación de otros servios

```python
df = pd.read_excel('Data/Can2019.xlsx', sheet_name='VAB por sector', engine='openpyxl', dtype={'CÓDIGO PROVINCIA': np.str, 
                                                                                     'CÓDIGO CANTÓN': np.str}
                  )
df.head(10)
```

```python
# Load maps

mapa = gpd.read_file('mapa_ecuador/', layer='ecu_admbnda_adm2_inec_20190724')
mapa.head(10)
```

```python
# En este caso, es conveniente unir las bases por códigos de cantón para no perder información. 
# Hay que hacer un corte del strung para eso.

mapa['Cod_cant'] = mapa['ADM2_PCODE'].str.slice(start=2) # Los códigos de cantón están contenidos en ADM2_ES
```

```python
df = mapa.merge(df, left_on='Cod_cant', right_on='CÓDIGO CANTÓN', suffixes=('_left', False))
```

```python
df.columns
```

# Dibujar mapas

```python
df_wgal = df[df['ADM1_ES'] != 'Galapagos']
```

```python
import matplotlib.pyplot as plt
```

# Sin Galapagos

```python
df_wgal = df_wgal[['Shape_Leng', 'Shape_Area', 'geometry', 'Servicios', 'VAB Economía total', 'Consumo intermedio']]
```

```python
# Hay que tener en cuenta que el VAB a precios corrientes es el PIB a precios corrientes. 
# Es decir, el VAB total para cada cada cantón es el PIB de cada cantón el dólares del 2019.
# Por desgracia, no hay información del PIB a precios constantes. Por lo que, no se puede obtener porcentajes
# pues sería una comparación inadecuada.
```

```python
# Setting , like decimal separator

pd.options.display.float_format = '{:,.2f}'.format
```

```python
# Crear ejes con contenedor y figura personalizando tamañp

fig, ax = plt.subplots(figsize=(10, 10))
 
# Mostrar el mapa finalizado

df_wgal.plot(column='Servicios',
        ax=ax, zorder=4,
       )

# Remover cuadros

ax.axis('off')


# TITLES

ax.set_title('VAB Servicios, 2019', 
             fontdict={'fontsize': '25', 'fontweight' : '3'})

# Create an annotation for the data source

ax.annotate('Fuente: BCE y OCHA\nAutor: Michael Peñafiel',
            xy=(0.1, .08), 
            xycoords='figure fraction', 
            horizontalalignment='left', 
            verticalalignment='top', 
            fontsize=15, 
            color='#555555'
           )

#Definition of the graphhical bar

bar = plt.cm.ScalarMappable(norm=plt.Normalize(vmin = df_wgal['Servicios'].min(), 
                                               vmax=df_wgal['Servicios'].max()
                                              )
                           )
#It is declared a list where the values of the bar will be saved

bar._A = []

#Location and size of the bar

cax = plt.axes([0.85, 0.15, 0.03, 0.7]) #[xcoord, ycoord, width, long]

# Plot of the graphical bar with the upper variables (bar and cax)

cbar = fig.colorbar(bar, cax=cax)

#Bar title

cbar.set_label('Servicios (miles de dólares)', fontsize=15)

# Guardar imagen 

plt.savefig('Graphs/servicespercant.png', dpi=300)
```

# Galapagos

```python
# Seleccionar variables de interés
df_gal = df[df['ADM1_ES'] == 'Galapagos']
df_gal = df_gal[['Shape_Leng', 'Shape_Area', 'geometry', 'Servicios', 'VAB Economía total', 'Consumo intermedio', 'PIB']]
```

```python
# Crear ejes con contenedor y figura personalizando tamañp

fig, ax = plt.subplots(figsize=(10, 10))
 
# Mostrar el mapa finalizado

df_gal.plot(column='Servicios',
            ax=ax, zorder=4
           )

# Remover cuadros

ax.axis('off')


# TITLES

ax.set_title('VAB Servicios, 2019', 
             fontdict={'fontsize': '25', 'fontweight' : '3'})

# Create an annotation for the data source

ax.annotate('Fuente: BCE y OCHA\nAutor: Michael Peñafiel',
            xy=(0.1, .08), 
            xycoords='figure fraction', 
            horizontalalignment='left', 
            verticalalignment='top', 
            fontsize=15, 
            color='#555555'
           )

#Definition of the graphhical bar

bar = plt.cm.ScalarMappable(norm=plt.Normalize(vmin=df_wgal['Servicios'].min(), # Para que cuadre con la escala del mapa sin galapagos
                                               vmax=df_wgal['Servicios'].max()
                                              )
                           )
#It is declared a list where the values of the bar will be saved

bar._A = []

#Location and size of the bar

cax = plt.axes([0.85, 0.15, 0.03, 0.7]) #[xcoord, ycoord, width, long]

# Plot of the graphical bar with the upper variables (bar and cax)

cbar = fig.colorbar(bar, cax=cax)

#Bar title

cbar.set_label('Servicios (miles de dólares)', fontsize=12)

# Guardar imagen 

plt.savefig('Graphs/servicesgal.png', dpi=300)
```
